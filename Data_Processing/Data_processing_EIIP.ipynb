```python
import numpy as np
import pandas as pd
import h5py as h5
```

```python
# Onehot encoding for the 4 nucleotides, 5th column is for EIIP conversion
Nuc_A = [1, 0, 0, 0, 0.1260]
Nuc_U = [0, 0, 1, 0, 0.1335]
Nuc_G = [0, 1, 0, 0, 0.0806]
Nuc_C = [0, 0, 0, 1, 0.1340]
Nuc_N = [0, 0, 0, 0, 0]
```

```python
file_path = 'Original_Data/Train_0.fa'
count = 0
Encoded_data = []
Encoded_label = []
```

```python
for line in open(file_path):
    len(line)
    if(line[0] == '>'):
        print(line.split()[-1])
        if(line.split()[-1] == '5S_rRNA'):
            Encoded_label.append(0)
        if(line.split()[-1] == '5_8S_rRNA'):
            Encoded_label.append(1)
        if (line.split()[-1] == 'tRNA'):
            Encoded_label.append(2)
        if (line.split()[-1] == 'ribozyme'):
            Encoded_label.append(3)
        if(line.split()[-1] == 'CD-box'):
            Encoded_label.append(4)
        if(line.split()[-1] == 'miRNA'):
            Encoded_label.append(5)
        if (line.split()[-1] == 'Intron_gpI'):
            Encoded_label.append(6)
        if (line.split()[-1] == 'Intron_gpII'):
            Encoded_label.append(7)
        if (line.split()[-1] == 'HACA-box'):
            Encoded_label.append(8)
        if(line.split()[-1] == 'riboswitch'):
            Encoded_label.append(9)
        if(line.split()[-1] == 'IRES'):
            Encoded_label.append(10)
        if (line.split()[-1] == 'leader'):
            Encoded_label.append(11)
        if (line.split()[-1] == 'scaRNA'):
            Encoded_label.append(12)
    else:
        Tem_List = []
        for i in range(750):
            if(i < len(line)-1):
                if(line[i] == 'A' or line[i] == 'a'):
                    Tem_List.append(Nuc_A)
                elif (line[i] == 'T' or line[i] == 't'):
                    Tem_List.append(Nuc_U)
                elif (line[i] == 'C' or line[i] == 'c'):
                    Tem_List.append(Nuc_C)
                elif (line[i] == 'G' or line[i] == 'g'):
                    Tem_List.append(Nuc_G)
                else:
                    Tem_List.append(Nuc_N)
            else:
                Tem_List.append(Nuc_N)
        Tem_List = np.array(Tem_List)
        Encoded_data.append(Tem_List)
```

```python
# Saving encoded data from original sequence data
Encoded_data = np.array(Encoded_data)
Encoded_label = np.array(Encoded_label)
f = h5.File("EIIP_data/Train_eiip", 'w') 
f.create_dataset('Train_Data', data=Encoded_data)
f.create_dataset('Label', data=Encoded_label)
f.close()
```
